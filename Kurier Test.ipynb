```python
from inca import Inca
myinca=Inca()
data=myinca.rssscrapers.kurier(save=False)
```

```python
for article in data:
    for k,v in article.items():
        if k=="title":
            print(v,"\n")
        elif k=="teaser":
            print(v,"\n")
        elif k=="text":
            print(v,"\n")
```
