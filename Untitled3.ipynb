```python
from inca import Inca
myinca=Inca()
```

```python
data=myinca.rssscrapers.thetelegraph(save=False)
```

```python
for article in data:
    for k,v in article.items():
        if k=="title":
            print(v,"\n")
        elif k=="category":
            print(v,"\n")
        elif k=="byline":
            print(v,"\n")
        elif k=="source":
            print(v,"\n")
        elif k=="text":
            print(v,"\n")
        elif k=="teaser":
            print(v,"\n")
        elif k=="authors":
            print(v,"\n")
```
